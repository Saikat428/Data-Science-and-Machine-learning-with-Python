# Normalization

```python
from sklearn.preprocessing import MinMaxScaler # import minmaxscaler function for normalization
```

```python
m = MinMaxScaler(feature_range=(0,1)) #create object of minmaxscaler funtion including range
```

```python
ms = m.fit(df[['Marketing Spend']]) #fit features in object and stored in ms variable
ms
```

```python
df.head() #see head of the dataframe
```

```python
df['Marketing Spend'] = ms.transform(df[['Marketing Spend']]) #transform a feature 
```

```python
ms #see the function name
```

```python
df.head() # see the transform feature 
```

```python
df['Administration'] = ms.fit_transform(df[['Administration']]) #here we do fit and transform at a time
```

```python
df.head()  # see the transform feature 
```

```python
df['Transport'] = ms.fit_transform(df[['Transport']]) #here we do fit and transform at a time
```

```python
df.head() # see the transform feature 
```
