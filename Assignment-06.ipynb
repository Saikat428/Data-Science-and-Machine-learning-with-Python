# Ordinal Encoder

```python
df7.head()#see the head of the dataset
```

```python
df7.region.unique() # see the unique value of the the feature region
```

```python
from sklearn.preprocessing import OrdinalEncoder #importing the ordinalencoder
```

```python
region=['southwest', 'southeast', 'northwest', 'northeast'] #created a list with value of region
```

```python
ordinal = OrdinalEncoder(categories=[region])# created object of the label ordinalencoder
encoded = ordinal.fit_transform(df7[['region']]) #fit and transform region
```

```python
encoded #see the encoded value
```

```python
new_region = pd.DataFrame(encoded, columns=['region']) #creaate new dataframe with encoded value
new_region.head() #see the new_region.head
```

```python
df7.head()#see the head of the dataset
```

```python
df7.region = new_region #store new_region in df7.region
df7.head()#see the head of the dataset
```

```python
df8.head()#see the head of the dataset
```

```python
cols = ['region']
for col in cols:
    unique = df8[col].unique()
    df8[col] = OrdinalEncoder(categories=[unique]).fit_transform(df8[[col]]) #fit and transform unique value of region 
```
